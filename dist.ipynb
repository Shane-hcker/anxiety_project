```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import scienceplots


plt.style.use(['science', 'notebook', 'grid'])
```

```python
anxiety_list = np.array([44, 46, 48, 49, 53, 54, 54, 55, 56, 56, 57, 58, 59, 61, 61, 62, 63,
       68, 69, 69, 72, 74, 74, 74, 76, 81, 81, 87, 89, 93])

s_hour = np.array([3, 5, 6, 8, 6.5, 4.5, 5, 11, 3.5, 5, 5.5, 12, 5, 4, 7.5, 13, 9, 10, 7, 
                   10, 8.5, 5, 3, 7, 6.5, 11, 10, 5.5, 4, 8.5])
s_hour.sort()
```

```python
fig1 = plt.figure(figsize=(8, 5))
ax1 = fig1.subplots()
ax1.set_title('Distribution of Anxiety Scores')

ax1.set_xlabel('Anxiety Scores(out of 100)')
ax1.set_ylim(0, 0.055)

bar1, sigma1 = norm.fit(anxiety_list)
ax1.hist(anxiety_list, color='green', density=True, alpha=0.7, label='Frequency Density Histogram')

pdf1 = norm.pdf(np.linspace(0, 100), bar1, sigma1)
ax1.plot(np.linspace(20, 100), pdf1, 'red', label='Fit Distribution')  # probability density function
ax1.legend(loc='upper right', fontsize=8)
```

```python
fig2 = plt.figure(figsize=(8, 5))
ax2 = fig2.subplots()
ax2.set_title('Distribution of Studying Hours')

ax2.set_xlabel('Study Hours(H)')

bar2, sigma2 = norm.fit(s_hour)
ax2.hist(s_hour, color='blue', density=True, alpha=0.5, label='Frequency Density Histogram')

pdf2 = norm.pdf(np.linspace(0, 20), bar2, sigma2)
ax2.plot(np.linspace(0, 20), pdf2, 'red', label='Fit Distribution')  # probability density function
ax2.legend(loc='upper right', fontsize=8)
```

```python
print(f"STD of Studying Hours: {sigma2:.2f}\nSTD of Anxiety Score: {sigma1:.2f}")
```
